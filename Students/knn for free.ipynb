```python
n_samples = len(digits.images)
data = digits.images.reshape(n_samples,64)#ย่อ array 8*8 ให้เป็น 1มิติ 8*8=64
data.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier

training_data = data[:1000] #แบ่งข้อมูล training กับ test
training_label = digits.target[:1000]
testing_data = data[1000:]
testing_label = digits.target[1000:]

classifier = KNeighborsClassifier(n_neighbors = 5) # k = 5 หยิบมา5รูป ย่อมมีโอกาศผิด
classifier.fit(training_data,training_label)

predictions = classifier.predict(testing_data)
```

```python
n_correct=sum(predictions == testing_label)
n_correct/len(testing_label)
```

```python
#row = round(((len(testing_label)-n_correct)/6),0)+1
#column = 6
#row
```

```python
#confusion matrix
cm = metrics.confusion_matrix(testing_label,predictions)
plt.imshow(cm)
cm.shape
```

```python
for i in range(len(predictions)):
    if(predictions[i] != testing_label[i]): 
        plt.imshow(digits.images[1000+i],cmap='gray_r')
        plt.ylabel(testing_label[i])  
        plt.xlabel(predictions[i], fontsize=18)   
        plt.show()
```
